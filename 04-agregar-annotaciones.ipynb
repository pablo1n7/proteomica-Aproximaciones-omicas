```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
data_protein = pd.read_csv('03_imputados_protein.groups.tsv', sep='\t', index_col=0)
```

```python
df_anotaciones = pd.read_csv('Cupriavidus_necator_H16_proteome_annotations_plus_C1.txt', sep='\t', index_col=0)
```

```python
df_final = pd.merge(
    data_protein, 
    df_anotaciones, 
    left_index=True,    # Usa el índice de data_protein
    right_on='Entry',   # Usa la columna 'Entry' de df_anotaciones
)
```
